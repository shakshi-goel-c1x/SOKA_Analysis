```python
category = new.apply(lambda x:x['UUID'].unique())
category = pd.DataFrame(category)
category.columns = ['UUID']
print(category)
```

```python
for index,row in category.iterrows():
    print(index)
    print(row)
    for i in row:
        print(i)
    
```

```python
top = {}
count = 0
for index,row in category.iterrows():
    top.update({index:0})
    for i in row:
        c = len(i)
    top[index]=c
    count = count+c
print(top)
#print(count)
```

```python
topCategory = [ [k,v] for k, v in top.items() ]
topCategory.sort(key = lambda x: x[1], reverse=True)  
print(topCategory) 
```

```python
df = pd.DataFrame(topCategory)
df.columns = ['ID','Count']
print(len(df))
```

```python
cat = pd.read_csv("category_soka.csv")
cat = cat[['id','name']]
cat = pd.DataFrame(cat)
cat.columns = ['MCID','Name']
cat.head()
```

```python
cat.nunique()
```

```python
Meta = []
Sub = []
for i,j in df['ID']:
    Meta.append(i)
    Sub.append(j)
print(len(Meta))
print(len(Sub))
```

```python
f = pd.DataFrame(columns = ['MCID','CID','Count'])
f['MCID'] = Meta
f['CID'] = Sub
f['Count'] = df['Count']
f
```

```python
result = pd.merge(f, cat, on='MCID')
result.head()
```

```python
catC = cat
catC.columns = ['CID','Name']
```

```python
topL = pd.merge(result, catC, on='CID')
topL.head()
```

```python
topL['Name'] = topL['Name_x'].str.cat(topL['Name_y'],sep="-")
topL.head()
```

```python
topL = topL.sort_values(by = 'Count', ascending = False)
topL
```

```python
MetaSub = pd.DataFrame(columns = ['Meta-Sub','Count'])
MetaSub['Meta-Sub'] = topL['Name']
MetaSub['Count'] = topL['Count']
MetaSub.head()
```

```python
exp = MetaSub.head(10)
exp
```

```python
f, ax = plt.subplots(figsize=(15, 8)) 
y_pos = np.arange(10)
plt.bar(y_pos, exp['Count'])
plt.xticks(y_pos, exp['Meta-Sub'],fontsize=15, rotation = 90)
plt.yticks(fontsize=15)
plt.xlabel('MetaCategory-SubCategory Name',fontsize=15)
plt.ylabel('Number of Unique UUIDs matched',fontsize=15)
plt.title('Top 10 MetaCategory-SubCategory',fontsize=20)
plt.show()
```
